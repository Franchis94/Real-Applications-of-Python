```python
from bokeh.plotting import figure
from bokeh.io import output_file,show

f=figure()

x=[1,4,5]
y=[2,3,6]

output_file("line.html")

f.triangle(x,y,size=[10,20,30],line_width=2)
show(f)
```

```python
from bokeh.plotting import figure
from bokeh.io import output_file,show
import pandas

f=figure(plot_width=500,plot_height=600,tools='pan')
f.xaxis.axis_label="Year"
f.yaxis.axis_label="Engineering"
f.title.text="Bachelors graph"
f.title.text_font="times"
f.title.text_font_style="bold"

df=pandas.read_csv("bachelors.csv")
x=df["Year"]
y=df["Engineering"]

output_file("bachelors_USA.html")

f.line(x,y)
show(f)
```

```python
from bokeh.plotting import figure
from bokeh.io import output_file,show
import pandas

f=figure(plot_width=500,plot_height=600,tools='pan')
f.xaxis.axis_label="Temperature deg. C"
f.yaxis.axis_label="Pressure (hPa)"
f.title.text="Weather graph"
f.title.text_font="times"
f.title.text_font_style="bold"

df=pandas.read_excel("weather.xlsx",sheetname=0)
x=df["Temperature"]
y=df["Pressure"]

output_file("Weather_USA.html")

f.circle(x,y)
show(f)
```

```python
from bokeh.plotting import figure
from bokeh.io import output_file,show
import pandas

f=figure(width=500,height=300,x_axis_type="datetime")
df=pandas.read_csv("adbe.csv",parse_dates=["Date"])

f.line(df["Date"],df["Close"],color='Orange',alpha=0.5)
output_file("Unscattered.html")
show(f)

```
